```python
import requests
from bs4 import BeautifulSoup
website_url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
```

```python
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
import pandas as pd
import numpy as np
my_table = soup.find('table', {'class': 'wikitable sortable'})

column_names = ['Postalcode','Borough','Neighborhood']
df = pd.DataFrame(columns = column_names)
postcode = 0
borough = 0
neighborhood = 0

for tr in my_table.find_all('tr'):
    i=0
    for td in tr.find_all('td'):
        if i==0:
            postcode = td.text
            i = i + 1
        elif i==1:
            borough=td.text
            i = i+1
        elif i==2:
            neighborhood = td.text.strip('\n').replace(']','')
            df = df.append({'Postalcode': postcode,'Borough': borough,'Neighborhood': neighborhood},ignore_index=True)

df = df[df.Borough!='Not assigned']
df = df[df.Borough!= 0]
df.reset_index(drop = True, inplace = True)
i = 0
for i in range(0,df.shape[0]):
    if df.iloc[i][2] == 'Not assigned':
        df.iloc[i][2] = df.iloc[i][1]
        i = i+1
                                 
my_df = df.groupby(['Postalcode','Borough'])['Neighborhood'].apply(', '.join).reset_index()
my_df
```

```python
!wget -q -O 'Toronto_long_lat_data.csv'  http://cocl.us/Geospatial_data
df_lon_lat = pd.read_csv('Toronto_long_lat_data.csv')
df_lon_lat.head()
```

```python
df_lon_lat.columns=['Postalcode','Latitude','Longitude']
```

```python
Toronto_df = pd.merge(my_df,
                 df_lon_lat[['Postalcode','Latitude', 'Longitude']],
                 on='Postalcode')
Toronto_df
```

```python
from geopy.geocoders import Nominatim

import matplotlib.cm as cm
import matplotlib.colors as colors

!conda install -c conda-forge folium=0.5.0 --yes 
import folium 

print('Libraries imported.')
```

```python
geolocator = Nominatim(user_agent="Toronto")
location = geolocator.geocode("Toronto")
latitude_toronto = location.latitude
longitude_toronto = location.longitude

longitude_toronto
latitude_toronto
```

```python
map_toronto = folium.Map(location=[latitude_toronto, longitude_toronto], zoom_start=10)

for lat, lng, borough, Neighbourhood in zip(Toronto_df['Latitude'], Toronto_df['Longitude'], Toronto_df['Borough'], Toronto_df['Neighborhood']):
    label = '{}, {}'.format(Neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
